```python
# Load data into SQL Server by creating new database Sales not under master database
import sqlalchemy as sal
from sqlalchemy import text
```

```python
# Create an engine connected to the master database
engine = sal.create_engine('mssql://Rajat\\SQLEXPRESS/master?driver=ODBC+DRIVER+17+FOR+SQL+SERVER')
conn = engine.connect().execution_options(isolation_level="AUTOCOMMIT")

# Create the Sales database if it does not exist
conn.execute(text("""
    IF NOT EXISTS (SELECT name FROM sys.databases WHERE name = 'Sales')
    BEGIN
        CREATE DATABASE Sales
    END
"""))

# Close the connection to master
conn.close()
```

```python
# Now connect to the new 'Sales' database
engine = sal.create_engine('mssql://Rajat\\SQLEXPRESS/Sales?driver=ODBC+DRIVER+17+FOR+SQL+SERVER')
conn = engine.connect()
print("Connected to the Sales database!")

#Explanation:
#execution_options(isolation_level="AUTOCOMMIT"): This is needed because CREATE DATABASE cannot be inside a transaction.

#The query now uses BEGIN...END to properly handle conditional creation.

# We then close the connection to master and reconnect to Sales.
```
